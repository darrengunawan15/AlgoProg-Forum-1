```python
#Forum Task 1

lloyd = {
    "name": "Lloyd",
    "homework": [90.0,97.0,75.0,92.0],
    "quizzes": [88.0,40.0,94.0],
    "tests": [75.0,90.0]
}

alice = {
    "name": "Alice",
    "homework": [100.0, 92.0, 98.0, 100.0],
    "quizzes": [82.0, 83.0, 91.0],
    "tests": [89.0, 97.0]
}

tyler = {
    "name": "Tyler",
    "homework": [0.0, 87.0, 75.0, 22.0],
    "quizzes": [0.0, 75.0, 78.0],
    "tests": [100.0, 100.0]
}

students = [lloyd, alice, tyler]

for student in students:
    print(f"Student's name         : {student['name']}")
    print(f"Student's homework     : {student['homework']}")
    print(f"Student's quizzes      : {student['quizzes']}")
    print(f"Student's tests        : {student['tests']}")
    print(" ")
    
def average(numbers):
    total = float(sum(numbers))
    avg = total / len(numbers)
    return avg

def get_average(student):
    homework = average(student['homework'])
    quizzes = average(student['quizzes'])
    tests = average(student['tests'])
    totalAvg = (homework*0.1) + (quizzes*0.3) + (tests*0.6)
    return totalAvg

def get_letter_grade(score):
    if score >= 0 and score <= 100:
        if score >= 90:
            return "A"
        elif score >= 80:
            return "B"
        elif score >= 70:
            return "C"
        elif score >= 60:
            return "D"
        else:
            return "F"
    else:
        return "Invalid Score"

# print(f"Lloyd's letter grade   : {get_letter_grade(get_average(lloyd))}") #80.55 => B
    
def get_class_average(students):
    results = []
    for student in students:
        results.append(get_average(student))
    avg = average(results)
    return avg
    
print(f"Class Average                : {get_class_average(students)}")
print(f"Class's Average (Rounded)    : {round(get_class_average(students), 2)}")
print(f"Class's Average letter grade : {get_letter_grade(get_class_average(students))}")
```

```python
#Forum Task 3

days_of_the_week = ("Sunday", "Monday", "Tuesday", "Wednesday", "Thursday", "Friday", "Saturday")

data = []

for day in days_of_the_week:
    day_sales = eval(input(f"Enter the sales for {day}: "))
    
    while day_sales <= 0:
        day_sales = eval(input(f"Input was invalid. Re-enter the sales for {day}: "))
    else:
        data.append(day_sales)

print(" ")
print(f"The total sales is          : ${sum(data):,.2f}")
print(f"The minimum sale amount was : ${min(data):,.2f}")
print(f"The maximum sale amount was : ${max(data):,.2f}")
```
